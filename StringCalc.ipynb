```python
import re


char = "(3+2+8)*41-(85/7445)-(58/69)+8"
count = 0
num = {}

result = re.finditer(r'\(.*?\)',char)
for item in result:
    num[count]=item.group(0)
    count+=1
print(num)

n1 = num[0]
print(n1)

def plus(num):
    if '+' in num:
        pos = num.find('+')
        n1=num[pos-1]
    #     print(n1)
        n2=num[pos+1]
    #     print(n2)
        result=float(n1)+float(n2)
    #     print(result)
        num=num.replace(num[:pos+2],str(result))
        return(num)
    
n1=plus(n1[1:-1])
n1=plus(n1)
print(n1)
char=char.replace(num[0],n1)
print(char)

def div(num):
    if '/' in num:
        pos = num.find('/')
        n1=num[pos-1]
    #     print(n1)
        n2=num[pos+1]
    #     print(n2)
        result=float(n1)/float(n2)
    #     print(result)
        num=num.replace(num[:pos+2],str(result))
        return(num)

n2 = num[1]
print(n2)
n2 = div(n2[1:-1])
char =char.replace(num[1],n2)
print(char)

n3 = num[2]
print(n3)
n3 = div(n3[1:-1])
char = char.replace(num[2],n3)
print(char)

def mul(num):
    pos = num.find('*')
    n1=num[pos-1]
#     print(n1)
    n2=num[pos+1]
#     print(n2)
    result=float(n1)*float(n2)
#     print(result)
    num=num.replace(num[:pos+2],str(result))
    return(num)

char = mul(char)
print(char)

def subt(num):
    pos = num.find('-')
    n1=num[pos-1]
#     print(n1)
    n2=num[pos+1]
#     print(n2)
    result=float(n1)-float(n2)
#     print(result)
    num=num.replace(num[:pos+2],str(result))
    return(num)

char = subt(char)
print(char)
char = subt(char)
print(char)
char = plus(char)
print(char)
```
